## Obteniendo Returns desde Precios


```python
R = X.pct_change()[1:]
print (R)
```

Graficando la distribucion en un histograma

```python
plt.hist(R, bins=20)
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['MSFT Returns']);
```

```python
np.mean(R)
```

```python
np.std(R)
```

Generando datos de una distribucin normal utilizando las estadsticas que estimamos a partir de las declaraciones de Microsoft. Vemos que hay buenas razones para sospechar que las devoluciones de Microsoft pueden no ser normales, ya que la distribucin normal resultante parece muy diferente.

```python
plt.hist(np.random.normal(np.mean(R), np.std(R), 10000), bins=20)
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['Normally Distributed Returns']);
```

Pandas tiene algunas buenas herramientas que nos permiten generar estadsticas progresivas. Aqu hay un ejemplo. Observe cmo no hay un promedio mvil durante los primeros 60 das, ya que no tenemos 60 das de datos para generar la estadstica.

```python
# Take the average of the last 60 days at each timepoint.
MAVG = pd.rolling_mean(X, window=60)
plt.plot(X.index, X.values)
plt.plot(MAVG.index, MAVG.values)
plt.ylabel('Price')
plt.legend(['MSFT', '60-day MAVG']);
```
