## Programa iterativo que calcula el n termino de la  la serie de Fibonacci en Python

Retorna el tiempo de ejecución grafica el tiempo de ejecución en función de n 

```python
import time
import numpy as np
import pylab as pl
#from __future__ import unicode_literals
%matplotlib inline

start_time=0

n = int(input())
def fibonacci(n):
    
    f=0
    if(n==0):
        return 0
    else:
        if(n==1):
            return 1
        else:
            a=0
            b=1
            for i in range (2,n+1):
                f=a+b
                a=b
                b=f
    return f

def tiempo(n):  
    return (n)
    


tiempos = []
for i in range(0,n+1):
    start = time.time()
    a = fibonacci(i)
    tiempos.append(time.time()-start) 


print(fibonacci(n))
print(tiempos[n]," segundos")

n=np.arange(0,n+1)
pl.ylabel("Tiempo de ejecucion")
pl.xlabel('n')
pl.plot(n,tiempos)
```
