```python
import numpy as np
import time
import pylab as pl
n= 1000
tiempos=[]
nn=0
aux=0
num=[]
count=10
for i in range(2,n+2):
    start_time = time.clock()
    a= np.ones((i, i))
    b= np.ones((i,i))
    b=b+1
    np.dot(a,b)
    #print time.clock()-start_time
    if nn<9:
        aux=aux+(time.clock()-start_time)
        nn=nn+1
    else:
        aux=aux+(time.clock()-start_time)
        tiempos.append(aux/10)
        #print(aux/10)
        num.append(count)
        count=count+10
        aux=0
        nn=0

#num.append(0)
pl.ylabel("Tiempo de ejecucion")
pl.xlabel('n')
pl.plot(num,tiempos)
```
